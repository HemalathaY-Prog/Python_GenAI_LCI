Inheritance

```python
# Parent class
class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    return "I am an animal."

# Child class inheriting from Animal
class Dog(Animal):
  def speak(self):
    return "Woof! My name is " + self.name

# Creating an instance of Dog
my_dog = Dog("Buddy")
print(my_dog.speak()) # Output: Woof! My name is Buddy
```

```python
#multiple inheritance
# Base class 1
class Father:
  def __init__(self):
    self.surname = "Smith"

  def print_surname(self):
    print("Surname:", self.surname)

# Base class 2
class Mother:
  def __init__(self):
    self.firstname = "Anna"

  def print_firstname(self):
    print("Firstname:", self.firstname)

# Child class inheriting from both Father and Mother
class Child(Father, Mother):
  def __init__(self):
    # Calling constructors of parent classes
    Father.__init__(self)
    Mother.__init__(self)

  def print_fullname(self):
    print("Fullname:", self.firstname, self.surname)

# Creating an instance of Child
my_child = Child()
my_child.print_fullname() # Output: Fullname: Anna Smith
```

Diamond inheritance in python
-----------------------------
It is a classic issue in OOP that occurs when a class inherits from two classes that both inherit from a single ancestor.

This can create ambiquity in the inheritance hierarchy,particualrly with reagrd to which ancestors method should be invloked.

In python,this problem resolved through the C3 linearization algorithm.This alogorithm provides a specific method resolution order(MRO) that python follows to determine which method to use in the presence of multiple inheritance.The MRO ensures that each class is only visited once,and it preserves the order in which the classes are specified in the inheritance list.

```python
class A:
  def show(self):
    print("I am from class A")

class B(A):
  def show(self):
    print("I am from class B")

class C(A):
  def show(self):
    print("I am from class C")

class D(B, C):
  pass

# Create an instance of D
d_instance = D()
d_instance.show() # This will print "I am from class B"
```

In the above example,D inherits from B and C which in turn inherits from A.when you call the show method on an instance of D,python will use the MRO to determine that B's method should be invoked before C's bcz B is listed first in the inheritance tuple of D.

We can view the MRO of a class by using __mro__ attribute or the mro() method

```python
print(D.__mro__)
# or
print(D.mro())

#This will ouput the the order in which python will search for methods to invoke,which helps in understanding how the diamond problem is resolved
```

Remember that while multiple inheritance and the diamond problem can be managed in pythonits often best to keep inheritance hierrachys as simple and unambiguious as possible.if the hierarchy becomes too complex it  might be worth considering alternative design patterns,such as composition.

```python
#Multilevel inheritance
# Base class
class Grandparent:
  def __init__(self):
    self.grandparent_name = "Grandpa"

  def show_grandparent(self):
    print("Grandparent:", self.grandparent_name)

# Intermediate class
class Parent(Grandparent):
  def __init__(self):
    super().__init__() # Call the constructor of Grandparent
    self.parent_name = "Dad"

  def show_parent(self):
    print("Parent:", self.parent_name)

# Derived class
class Child(Parent):
  def __init__(self):
    super().__init__() # Call the constructor of Parent
    self.child_name = "Child"

  def show_child(self):
    print("Child:", self.child_name)

# Creating an instance of Child
my_child = Child()
my_child.show_grandparent() # Output: Grandparent: Grandpa
my_child.show_parent() # Output: Parent: Dad
my_child.show_child() # Output: Child: Child
```

Super()
=======
It is used to give access to methods from a parent class from within subclass.
This particulary useful when you want to extend or modify the functionality of inherited methods.


It is used to call the constructor of the immediate parent class,ensuring that the initialization of the parent classes is handled properly.

```python
# Parent class
class Vehicle:
  def __init__(self, category):
    self.category = category

  def show_category(self):
    print(f"This is a {self.category} vehicle.")

# Child class
class Car(Vehicle):
  def __init__(self, category, model):
    super().__init__(category) # Call the __init__ of the parent class Vehicle
    self.model = model

  def show_car(self):
    # Call the show_category method from the parent class using super()
    super().show_category()
    print(f"The model of the car is {self.model}.")

# Creating an instance of Car
my_car = Car("land", "Toyota Corolla")
my_car.show_car()
```

Method overriding:
------------------
*It occurs when a subclass provides a specific implementation of a method that is already defined in its superclass.
*This allows the subclass to implement a behavior thats specific to it,even though a method with the same name exists in the superclass.   



```python
# Superclass
class Bird:
  def sing(self):
    print("Bird sings")

# Subclass
class Parrot(Bird):
  # Overriding the sing method
  def sing(self):
    print("Parrot sings beautifully")

# Creating an instance of Parrot
polly = Parrot()
polly.sing() # This will output: Parrot sings beautifully
```

In the above example ,the parraot class overrides the sing method of the bird class.when you call the sing method on an instance of parraot,the overridden method in the parrot class is executed,not the one in the bird class.

This is a fundamental concept on OOPs as it allows for polymorphism,where the same method can have different behaviors on which subclass is invoking it.

Operator Overloading:
----------------------
It allows the same operator to have different meanings based on the context.
Specifically, it means giving extended meaning beyond their predefined operational meaning.

For ex, the + operatoe can be used to add two integers ,join two strings,and merge twi lists bcz it is overloaded by the int class and str class.

```python
a = 10
b = 20

print (a+b)

a = "Hello "
b = "World"
print(a+b)
```

```python
#+ operator for custom class
class Point:
  def __init__(self,x,y):
    self.x=x
    self.y =y

  def __str__(self):
    return f"{self.x,self.y}"

 # operator + is overloaded
  def __add__(self,other):
    x = self.x + other.x
    y = self.y + other.y
    return Point(x,y)

  # operator * is overloaded
  def __mul__(self,other):
    x = self.x * other.x
    y = self.y * other.y
    return Point(x,y)

  def __eq__(self,other):
    return self.__dict__ ==other.__dict__
```

```python
p1 = Point(3,4)
p2 = Point(3,4)
print(p1)
print(p2)

p3 = p1 + p2
#p3 = "sam" + "joe"
print(p3)

p3 = p1 * p2
#p3 = "sam" + "joe"
print(p3)

print(p1 == p2) # True - Values of the objects
print(p1 is p2) # False - Address of the objects
```

Encapsulation and data hiding:
==============================
It is a fundamental concept in OOPs that involves bundling data(variables) and methods(functions) that operate on the data into a single unit or class.

Data hiding: is a part of encapsulation that restricts access to certain components of an object, which means that the object's internal state connot be directly accessed from outside the object; instead it can only be manipulated through its methods.

```python
class BankAccount:
  def __init__(self, account_number, balance):
    self.__account_number = account_number # Private attribute
    self.__balance = balance # Private attribute

  def deposit(self, amount):
    if amount > 0:
      self.__balance += amount
      print(f"Deposit successful. Balance: {self.__balance}")
    else:
      print("Invalid deposit amount.")

  def withdraw(self, amount):
    if 0 < amount <= self.__balance:
      self.__balance -= amount
      print(f"Withdrawal successful. Balance: {self.__balance}")
    else:
      print("Invalid withdrawal amount or insufficient funds.")

  def get_balance(self):
    return self.__balance

# Creating an instance of BankAccount
my_account = BankAccount("123456789", 1000)

# Accessing methods
my_account.deposit(500)
my_account.withdraw(200)
print("Current Balance:", my_account.get_balance())

# Trying to access private attributes directly will result in an AttributeError
print(my_account.__balance) # Uncommenting this line will raise an error
```

In this ex, __account_number and __balance are private attrs,indicated by the __ prefix,this means they can not be accessed directly from outside of the class.

the deposit and withdraw are public methods,allowing controlled interaction with the private data.attempting to access __balance directly will result an attribute error ,demonstrating data hiding.
